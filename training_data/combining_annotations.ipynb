## Finding sentences which were given three different labels!! ##

```python
three_different = []

# Iterate over the dataframe rows using iterrows()
for index, row in merged_all_clean.iterrows():
    values = [row['eric_category'], row['lucia_category'], row['sada_category']]
    if len(set(values)) == 3:
        three_different.append(index)

# Filter the dataframe based on the identified indices
df_three_different = merged_all_clean.loc[three_different]

df_three_different.dropna(subset=['eric_category', 'lucia_category', 'sada_category'], inplace=True)
```

## Finding sentences with at least two common labels ##

```python
common_value_indices_2 = []

for index, row in merged_all_clean.iterrows():
    values = [row['eric_category'], row['lucia_category'], row['sada_category']]
    values = [v for v in values if pd.notna(v)]
    if len(set(values)) < len(values):
        common_value_indices_2.append(index)
```

```python
len(common_value_indices_2)
```

```python
# Filter the dataframe based on the identified indices
two_way_cross_validated = merged_all_clean.loc[common_value_indices_2]

two_way_cross_validated.head()
```

```python
two_way_cross_validated.info()
```

```python
two_way_cross_validated.to_excel("2way_cross_validated.xlsx")
```

 ## Figuring out how many of the two-way cross-validated sentences have a NaN value as the third, or a disagreement label ##

```python
validated_no_nans = two_way_cross_validated.dropna(subset=['eric_category', 'lucia_category', 'sada_category'])
```

```python
len(validated_no_nans)

print("Number of two-way cross-validated with a disagreement label: " + str(len(validated_no_nans)))
print("Number of two_way cross-validated with a MISSING third label: " + str((len(common_value_indices_2)-len(validated_no_nans))))
```

```python
validated_no_nans.head()
```

```python
x = []

for index, row in validated_no_nans.iterrows():
    values = [row['eric_category'], row['lucia_category'], row['sada_category']]
    if len(set(values)) == 2:
        x.append(index)
```

```python
one_disagreement = validated_no_nans.loc[x]
```

```python
one_disagreement.head()
```

```python
one_disagreement.to_excel("one_disagreeing_label.xlsx")
```

## Getting the sentences with three-way agreement on the label ##

```python
common_value_indices_1 = []

for index, row in merged_all_clean.iterrows():
    values = [row['eric_category'], row['lucia_category'], row['sada_category']]
    if len(set(values)) <= 1:
        common_value_indices_1.append(index)
```

```python
len(common_value_indices_1)
```

```python
# Filter the dataframe based on the identified indices
three_way_cross_validated = merged_all_clean.loc[common_value_indices_1]

three_way_cross_validated.head()
```
